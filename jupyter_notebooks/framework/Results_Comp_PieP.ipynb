```python
colors1 = ["xkcd:forest green", "xkcd:burnt orange", "xkcd:cobalt"]
colors2 = ["xkcd:olive green", "xkcd:pumpkin", "xkcd:ocean blue"]
colors3 = ["xkcd:grey green", "xkcd:pale orange", "xkcd:grey blue"]
explode1 = ( 0, 0.0, 0.1)
explode2 = ( 0, 0.0, 0.1)
explode3 = ( 0, 0.0, 0.1)
```

```python
Fig1, axs = plt.subplots(nrows=1, ncols=3)
Fig1.set_size_inches( (18.0, 9.0) )
pn1 = axs[2].pie( mHSP2_B_P3, explode=explode3, labels=labels, colors=colors3, 
                autopct='%1.1f%%', shadow=True, startangle=140,
                textprops={'fontsize':12} )
axs[2].axis('equal')
axs[2].annotate("C", xy=(0.10,0.18), xycoords="axes fraction", fontsize=11.0,
              backgroundcolor="xkcd:light grey")
pn2 = axs[1].pie( mHSP2_P3, explode=explode2, labels=labels, colors=colors2, 
                autopct='%1.1f%%', shadow=True, startangle=140,
                textprops={'fontsize':12} )
axs[1].axis('equal')
axs[1].annotate("B", xy=(0.10,0.18), xycoords="axes fraction", fontsize=11.0,
              backgroundcolor="xkcd:light grey")
pn3 = axs[0].pie( MWB_P3, explode=explode1, labels=labels, colors=colors1, 
                autopct='%1.1f%%', shadow=True, startangle=140,
                textprops={'fontsize':12} )
axs[0].axis('equal')
axs[0].annotate("A", xy=(0.10,0.18), xycoords="axes fraction", fontsize=11.0,
              backgroundcolor="xkcd:light grey")
Fig1.savefig( OutSVG )
Fig1.savefig( OutPNG )
```
