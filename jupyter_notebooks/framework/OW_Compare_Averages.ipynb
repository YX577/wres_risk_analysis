# Comparison of Smoothed Day of the Year Averages

Fourier series smoothing was used to create day of the year average values for "other weather" parameters from each of the available data sets and time periods.

The available data sets and parameters are enumerated below.

- PRISM
    - Maximum daily air temperature
    - Average daily air temperature, provided as the average between the maximum and the minimum
    - Minimum daily air temperature
    - Relative humidity, calculated from dew point temperatuer and vapor pressure deficit
- CMIP5, LOCA downscaling
    - Maximum daily air temperature
    - Minimum daily air temperature
- CMIP5, BCCA downscaling
    - Maximum daily air temperature
    - Minimum daily air temperature

Climate periods:

1. Data period: 1981-2010
2. Projection period 1: 2011-2040
3. Projection period 2: 2041-2070
4. Projection period 3: 2071-2100
